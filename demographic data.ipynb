```python

# Calculate the total number of people in each country
total_people_by_country = df.groupby('native-country').size()

# Calculate the number of people earning >50K in each country
high_salary_by_country = df[df['salary'] == '>50K'].groupby('native-country').size()

# Calculate the percentage of people earning >50K in each country
percentage_high_salary_by_country = (high_salary_by_country / total_people_by_country) * 100

# Find the country with the highest percentage
highest_percentage_country = percentage_high_salary_by_country.idxmax()
highest_percentage = percentage_high_salary_by_country.max()

print(f"The country with the highest percentage of people earning >50K is {highest_percentage_country} with a percentage of {highest_percentage:.2f}%")
```
